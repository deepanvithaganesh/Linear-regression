```python
import numpy as np
import matplotlib.pyplot as plt
def gradient_descent(x,y):
    m1=b1=0
    n=len(x)
    intervals=1000
    learning_rate=0.01
    plt.scatter(x,y,color='blue')
    for i in range(intervals):
        y1=m1*x+b1
        plt.plot(x,y1,color='y',marker='+')
        md= -(2/n)*sum(x*(y-y1))
        bd= -(2/n)*sum(y-y1)
        m1=m1-learning_rate * md
        b1=b1-learning_rate * bd
        print ('m {}, b {}, cost {} iteration {}'.format(m1,b1,md, i))
        

a=np.array([1,2,3,4,5])
b=np.array([8,6,9,11,13])

gradient_descent(a,b)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)
X_train
```

```python
y=tr.predict(X_test)
y
```
